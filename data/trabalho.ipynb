```python
import pandas as  pd
from sqlalchemy import create_engine
data=pd.read_csv("project_files/G25_Food Delivery – Restaurants  Dishes with Delivery Partners_merged (1).csv")
```

```python
columns_to_check = ['restaurant_id', 'dish_id', 'partner_id']
data = data.drop_duplicates(subset=columns_to_check, keep='first').copy()
data.loc[:, 'id'] = range(1, len(data) + 1)
data
```

```python
from sqlalchemy import create_engine
columns_to_import = ['cuisine', 'restaurant_id', 'restaurant_name']
df_selected = data[columns_to_import]
database_path = 'Restaurant.db'  
engine = create_engine(f'sqlite:///{database_path}')
table_name = 'Restaurant'
df_selected.to_sql(table_name, con=engine, if_exists='replace', index=False) 
```
